## Step 1: Loading the IRIS data set from GitHub gists using JavaScript's fetch 
We can simply import JS functions in our Python function using `from js import ...`.
In this case, we will use JavaScript's fetch function to load the a .csv file containing the IRIS data set from GitHub. 
The file is then saved locally using Python's file writer. 

```plaintext
from js import fetch

async def _download_iris_data_set(filename):
    URL = f"https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv"
    res = await fetch(URL)
    text = await res.text()

    with open(filename, 'w') as f:
        f.write(text)
            
filename = "iris.csv"
print(f'Download {filename} from GitHub')
res = await _download_iris_data_set(filename)
```

```plaintext
def update_rcparams(highlights_color, facecolor, facecolor_alpha):
    plt.rcParams.update({
        "figure.facecolor":  f"{facecolor}{facecolor_alpha}",  # white with alpha = 10%
        "axes.facecolor":    f"{facecolor}{facecolor_alpha}",  # white with alpha = 10%
        "text.color": highlights_color,
        "axes.labelcolor": highlights_color,
        "axes.edgecolor": highlights_color,
        "xtick.color": highlights_color,
        "ytick.color": highlights_color,
    })
```
